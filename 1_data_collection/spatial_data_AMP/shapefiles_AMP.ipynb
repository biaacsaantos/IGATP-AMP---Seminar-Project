```python
# Lista de códigos extraídos
codigos_4dig = ['0104', '0107', '0109', '0113', '0116', '0119',
                '1304', '1306', '1308', '1310', '1312', '1313',
                '1314', '1315', '1316', '1317', '1318']

# Filtrar o DataFrame shape_CAOP_Freg
shape_CAOP_Freg_AMP = shape_CAOP_Freg[
    shape_CAOP_Freg["DICOFRE"].str[:4].isin(codigos_4dig)
]
```

```python
shape_CAOP_Freg_AMP
```

```python
# Contar o número de municípios distintos
numero_municipios = shape_CAOP_Freg_AMP["Municipio"].nunique()

# Obter a lista de municípios únicos, ordenada alfabeticamente
lista_municipios = sorted(shape_CAOP_Freg_AMP["Municipio"].unique())

# Mostrar o número de municípios e a lista
print(f"Número de municípios distintos: {numero_municipios}")
print("Lista de municípios:")
for municipio in lista_municipios:
    print("-", municipio)
```

```python
# Garantir que os DataFrames têm a mesma projeção
shape_CAOP_Conc = shape_CAOP_Conc.to_crs(shape_CAOP_Conc_AMP.crs)

# Fazer um spatial join para associar os concelhos da CAOP à AMP
shape_CAOP_Conc_AMP = gpd.sjoin(shape_CAOP_Conc, shape_CAOP_Conc_AMP, predicate="within")

# Visualizar o resultado
shape_CAOP_Conc_AMP.plot()
```

```python
shape_CAOP_Conc_AMP.columns
```

```python
# Obter os valores únicos da coluna 'DICO' do DataFrame shape_AMP
shape_CAOP_Conc_AMP['Municipio_left'].unique()
```

```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(7, 7))

# Fazer o gráfico do polígono único da AMP (fundo)
shape_CAOP_Conc_AMP.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)

# Fazer o gráfico dos limites dos municípios dentro da AMP (com linhas pretas)
shape_CAOP_Conc_AMP.plot(ax=ax, facecolor="none", edgecolor="black", linewidth=1)

# Adicionar os nomes dos municípios
for idx, row in shape_CAOP_Conc_AMP.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, row["Municipio_left"],
             fontsize=8, ha='center', color='black', bbox=dict(facecolor='white', edgecolor='none', alpha=0.7))

# Mostrar o mapa
plt.title("Área Metropolitana do Porto com Limites dos Municípios", fontsize=12)
plt.show()
```

```python
# Garantir que os DataFrames têm a mesma projeção
shape_CAOP_Freg = shape_CAOP_Freg.to_crs(shape_CAOP_Freg_AMP.crs)

# Fazer um spatial join para associar os concelhos da CAOP à AMP
shape_CAOP_Freg_AMP = gpd.sjoin(shape_CAOP_Freg, shape_CAOP_Freg_AMP, predicate="within")

# Visualizar o resultado
shape_CAOP_Freg_AMP.plot()
```

```python
# Exportar o GeoDataFrame dos concelhos da AMP
shape_CAOP_Conc_AMP.to_file("shape_CAOP_Conc_AMP.shp", driver="ESRI Shapefile")

# Exportar o GeoDataFrame das freguesias da AMP
shape_CAOP_Freg_AMP.to_file("shape_CAOP_Freg_AMP.shp", driver="ESRI Shapefile")
```
